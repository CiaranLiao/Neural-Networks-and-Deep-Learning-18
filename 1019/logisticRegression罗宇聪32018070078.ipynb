```python
#实现逻辑回归，创建模型预测结果
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#代价函数
def cost(theta,X,y):
    first=(-y)*np.log(sigmoid(X @ theta))
    second=(1-y)*np.log(1-sigmoid(X @ theta))
    return np.mean(first-second)
```

```python
#插入一项全是1的列
if 'Ones' not in data.columns:
    data.insert(0,'Ones',1)
#设置训练集目标集
X=data.iloc[:,:-1].values #转为数组形式，这个方法快被淘汰了，用.value
y=data.iloc[:,-1].values
theta=np.zeros(X.shape[1])#设置theta向量
```

```python
X.shape,theta.shape,y.shape
```

```python
cost(theta,X,y)
```

```python
#计算训练数据，标签，和一些参数theta的梯度
def gradient(theta,X,y,n):
    j=[]
    for i in range(n):
        temp=(np.dot(X.T,sigmoid(np.dot(X,theta)-y)))/(len(X))
        theta=temp
        j.append(cost(theta,X,y))
    return theta,j
n=100
theta,cost2=gradient(theta,X,y,n)
cost(theta,X,y)
#cost可视化
plt.figure()
list=[i for i in range(100)]
plt.plot(list,cost2)
plt.xlabel("Iterations")
plt.ylabel('cost')
plt.title('Error vs Training Epoch')
plt.show()
```

```python
import scipy.optimize as opt
```

```python
result=opt.fmin_tnc(func=cost,x0=theta,fprime=gradient,args=(X,y))
result
```

```python
#预测函数

def predict(theta,X):
    probability=sigmoid(X@theta)
    return[1 if x>=0.5 else 0 for x in probability]
final_theta=result[0]
predictions=predict(final_theta,X)
#predictions

#correct=[1 if a==b else 0 for (a,b) in zip(predictions,y)]
#accuracy=sum(correct)/len(X)
#accuracy
```
