```python
import numpy as np
import matplotlib.pyplot as plt

#-----------------计算代价函数-----------------------
def computeCost(X, y, theta):
    m = np.size(X[:,0])   #返回矩阵的元素个数
    J = 1/(2*m)*np.sum((np.dot(X,theta)-y)**2)
    return J

 #-------------------根据面积和卧室数量预测房价----------
#读取数据集中数据，第一列是面积数据，第二列是卧室数量，第三列是房价
data = np.loadtxt('D:\\mystyle\\人工神经网络\\ppt\\ex1data2.txt',delimiter=",",dtype="float")
m = np.size(data[:,0])

X = data[:,0:2]
y = data[:,2:3]

#----------------------均值归一化---------------------
mu = np.mean(X,0)   #对列求均值
sigma = np.std(X,0)
X_norm = np.divide(np.subtract(X,mu),sigma)
one = np.ones(m)   #添加第一列1
X_norm = np.insert(X_norm,0,values=one,axis=1)

# h(x)=theta1+theta2*x1+theta3*x2

#----------------------梯度下降-----------------------
alpha = 0.05
iterations = 100
theta = np.zeros((3,1));  #初始化theta
J_history = np.zeros((iterations,1))  #初始化代价函数值
for i in range(0,iterations):
    theta = theta - alpha/m*np.dot(X_norm.T,(np.dot(X_norm,theta)-y))
    J_history[i] = computeCost(X_norm,y,theta)
print(theta)
x_col = np.arange(0,iterations)
plt.plot(x_col,J_history,'-b')
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')
plt.show()
print(J_history)

#----------使用上述结果对[1650,3]的数据进行预测--------
test1 = [1,1650,3]
test1[1:3] = np.divide(np.subtract(test1[1:3],mu),sigma)
price = np.dot(test1,theta)
print(price)        #输出预测结果[292455.63375132]
```
