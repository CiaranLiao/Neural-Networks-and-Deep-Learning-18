```python
import math
import numpy as np
def sigmoid(z):
    return 1/(1+math.exp(-z))
print(sigmoid(0))
class Perceptor(object):
    def __init__(self,w,b):
        self.w=w
        self.b=b
    def __call__(self,x):
        return sigmoid(np.dot(x,self.w)+self.b)
pe=Perceptor([1,2,3],-4)
output=pe([1,0,1])
print(output)
output=pe([1,1,0])
print(output)
```
